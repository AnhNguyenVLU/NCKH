```python
sample = data.loc[data["ID"] == "197QT28220"]
dct = {
    "viewed":[],
    "submitted":[],
    "attempt started":[],
    "uploaded":[]
}
y = {}
x = []
```

```python
for event in sample["Event name"].unique().tolist():
    for key in dct.keys():
        if key in event:
            dct[key].append(event)
```

```python
sample = sample.astype({"Time":"datetime64"})
startDate = sample.sort_values(["Time"], ascending=True).iloc[0,0]
endDate = sample.sort_values(["Time"], ascending=True).iloc[-1,0]
during = (endDate - startDate).days //10
```

```python
while startDate <= endDate:
    time = startDate + datetime.timedelta(during)
    limitbyTime = sample.loc[(sample["Time"]>startDate) & (sample["Time"]<time)]
    for key, item in dct.items():
        result = 0
        for i in item:
            result += len(limitbyTime.loc[limitbyTime["Event name"]==i])
        if key in y:
            y[key].append(result)
        else:
            y[key] = [result]
    x.append(time)
    startDate = time
```

```python
# plt.plot(x, y["viewed"], linestyle="-", color="C0", label="viewd")
plt.plot(x, y["submitted"], linestyle="-", color="C1", label="submitted")
plt.plot(x, y["attempt started"], linestyle="-", color="C2", label = "attempt")
plt.plot(x, y["uploaded"], linestyle="-", color="C3", label = "uploaded")
plt.xticks(rotation=90)
plt.title("197QT28220 event")
plt.legend()
plt.show()
```
